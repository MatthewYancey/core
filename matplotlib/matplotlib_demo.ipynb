```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

### Style Sheets

```python
# lists the style sheets available
print(plt.style.available)
```

```python
# data for ploting
t = np.arange(0.0, 2.0, 0.01)
s1 = 1 + np.sin(2*np.pi*t)
s2 = 1 + np.cos(2*np.pi*t)
s3 = 2 + np.sin(2*np.pi*t)
s4 = 2 + np.cos(2*np.pi*t)


# sample line chart with default styling
plt.plot(t, s1)
plt.plot(t, s2)
plt.plot(t, s3)
plt.plot(t, s4)
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.show()
```

```python
# loads a style sheet
plt.style.use(['fivethirtyeight'])

plt.plot(t, s1)
plt.plot(t, s2)
plt.plot(t, s3)
plt.plot(t, s4)
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.show()
```

```python
# style sheets are located in 
# Anaconda3\Lib\site-packages\matplotlib\mpl-data\stylelib
# we can also load custom ones from our current working directory

# loads a style sheet
plt.style.use(['custom.mplstyle'])

plt.plot(t, s1)
plt.plot(t, s2)
plt.plot(t, s3)
plt.plot(t, s4)
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Simple Graph')
plt.show()
```

### Bubble graph

```python
# create data
x = [0, 2, 4, 6, 8, 10]
y1 = [0]*len(x)
s1 = [20*4**n for n in range(len(x))]

y2 = [i * 0.1 for i in x]
s2 = s1[::-1]

y3 = y2[::-1]
s3 = s1

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bubble Plot')
plt.scatter(x, y1, s = s1, alpha = 0.75)
plt.scatter(x, y2, s = s2, alpha = 0.75)
plt.show()
```
